```python
import pandas as pd
from datasets import load_dataset

# Load CSV as a Dataset
dataset = load_dataset("csv", data_files="imbalance.csv", split="train", encoding="utf-8", quotechar='"')

def get_teencode(ori, nor):
    ori_words = ori.split()
    nor_words = nor.split()
    teencode = [word for word in ori_words if word not in nor_words]
    return teencode

def get_instructions(ori, nor):
    rows = []

    # Template 1
    rows.append({
        "Instruction": "Hãy chuẩn hóa câu sau thành câu bình thường:",
        "Input": ori,
        "Answer": nor,
        "Task": "normalization"
    })

    # Template 2
    rows.append({
        "Instruction": "Cho cặp câu trong đó câu 1 là câu chứa các từ teencode và câu 2 là câu đã chuẩn hóa sai các từ teencode ở câu 1, hãy chuẩn hóa lại các từ bị chuẩn hóa sai",
        "Input": f"Câu 1: {ori}, Câu 2: {ori}",
        "Answer": f"Các từ bị chuẩn hóa sai: {get_teencode(ori, nor)}, sửa lại: {nor}",
        "Task": "detection"
    })

    # Template 3
    rows.append({
        "Instruction": "Cho biết các từ nào là teencode trong câu sau:",
        "Input": ori,
        "Answer": f"Các từ teencode: {get_teencode(ori, nor)}",
        "Task": "detection"
    })
    return rows

instructions = []
for ori, nor in zip (dataset['original'], dataset['normalized']):
    instructions.extend(get_instructions(ori, nor))

df = pd.DataFrame(instructions)

# Save to CSV
df.to_csv("instruction_imbalance.csv", index=False, encoding="utf-8")
```
