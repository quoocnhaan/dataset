```python
from datasets import load_dataset
from unidecode import unidecode
from datasets import concatenate_datasets

dataset = load_dataset("csv", data_files="synthetic_data.txt", delimiter=",")['train']
sampled = dataset.shuffle(seed=42).select(range(1000))
```

```python
def normalize(batch):
    return {
        'original': [unidecode(e) for e in batch['original']]
    }
sampled = sampled.map(normalize, batched=True)
```

```python
combined = concatenate_datasets([dataset, sampled])
combined.to_csv("synthetic_updated.csv", index=False)
```
