```python
import pandas as pd
from datasets import load_dataset, Dataset

# Load CSV files as Hugging Face datasets
data = load_dataset("csv", data_files="train_data.csv", split="train", encoding="utf-8", quotechar='"')

def split_data(batch):
    ori_filtered = []
    nor_filtered = []
    for ori, nor in zip (batch['original'], batch['normalized']):
        if len(ori.split()) != len(nor.split()):
            ori_filtered.append(ori)
            nor_filtered.append(nor)
    return {
        'original': ori_filtered,
        'normalized': nor_filtered
    }

# Save to CSV
new_data = data.map(split_data, batched=True)

new_data.to_csv('imbalance.csv')
```
