```python
from datasets import load_dataset

# Load the dataset
dataset = load_dataset("csv", data_files="viohs.csv")["train"]
```

```python
# Keep only one column
column = "sentence"
removed_columns = [c for c in dataset.column_names if c != column]
dataset = dataset.remove_columns(removed_columns)
dataset.to_csv('fb_cmt.csv')
```
