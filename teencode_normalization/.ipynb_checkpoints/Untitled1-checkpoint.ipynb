```python
from datasets import load_dataset

# Load your CSV
data_files = {
    "train": "normalized_train.csv",
}

dataset = load_dataset('csv', data_files=data_files)
```

```python
import re
import unicodedata

def normalize_text(text):
    # slang_dict = {
    #     'ntn': 'như thế này',
    #     'k': 'không',
    #     'ko': 'không',
    #     'kh': 'không',
    #     'kg': 'không',
    #     'j': 'gì',
    #     'cj': 'cái gì',
    #     'm': 'mày',
    #     'mk': 'mình',
    #     'mn': 'mọi người',
    #     'vl': 'vãi lồn',
    #     'vcl': 'vãi cái lồn',
    #     'đc': 'được',
    #     'dc': 'được',
    #     'thik': 'thích',
    #     'cmnl': 'con mẹ nó lời',
    #     'str': 'story',
    #     'ccj': 'con cặc gì',
    #     'vc': 'vãi cặc',
    #     'jz': 'gì vậy',
    #     'cfs': 'confesion',
    #     't3': 'thứ 3',
    #     '10đ': '10 điểm',
    #     '10tr': '10 triệu',
    #     'đcmm': 'địt con mẹ mày',
    #     'tttm': 'trung tâm thương mại',
    #     'tq': 'Trung Quốc',
    #     's': 'sao',
    #     'kmn': 'kệ mẹ nó',
    #     'c1': 'cấp 1',
    #     'vn': 'việt nam',
    #     'hnao': 'hôm nào',
    #     'mxh': 'mạng xã hội',
    #     'thuj': 'thôi',
    #     'ip': 'iphone',
    #     'kkk': 'haha',
    #     '7.5đ': '7.5 điểm',
    #     '90p': '90 phút',
    #     '3tr': '3 triệu',
    #     'hn': 'hà nội',
    #     '10tr200': '10 triệu 200',
    #     'kphai': 'không phải',
    #     'c2': 'cấp 2',
    #     'c3': 'cấp 3',
    #     '50tr': '50 triệu',
    #     'cp': 'couple',
    #     '7tr500': '7 triệu 500',
    #     '5ph': '5 phút',
    #     '15p': '15 phút',
    #     '30t': '30 tuổi',
    #     'bm': 'bố mẹ',
    #     'fl': 'follow',
    #     'any': 'anh người yêu',
    #     'uk': 'ừa',
    #     '10tr': '10 triệu',
    #     'fb': 'facebook',   
    #     'mqh': 'mối quan hệ',
    # }

    if not isinstance(text, str):
        return text

    # for slang, full in slang_dict.items():
    #     pattern = r'\b' + re.escape(slang) + r'\b'
    #     text = re.sub(pattern, full, text, flags=re.IGNORECASE)
    # 1. Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    #2. Normalize tone marks
    text = unicodedata.normalize('NFC', text)
    # 3. Remove punctuations
    text = re.sub(r'[^\w\s]', '', text)
    return text

dataset = dataset.map(lambda batch: {'original': [normalize_text(t) for t in batch['original']]}, batched=True)
dataset = dataset.map(lambda batch: {'normalized': [normalize_text(t) for t in batch['normalized']]}, batched=True)


dataset['train'].to_csv('normalized_train.csv')
dataset['test'].to_csv('normalized_test.csv')
dataset['validation'].to_csv('normalized_val.csv')
```
