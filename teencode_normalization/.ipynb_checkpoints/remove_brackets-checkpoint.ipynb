```python
from datasets import load_dataset
from datasets import concatenate_datasets


# Load both datasets
train_ds = load_dataset("csv", data_files="normalized_train_updated.csv")["train"]
synthetic_ds = load_dataset("csv", data_files="synthetic_updated.csv")["train"]
```

```python
combined = concatenate_datasets([train_ds, synthetic_ds])
combined.to_csv("train_data.csv", index=False)
```
