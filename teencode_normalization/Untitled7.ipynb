```python
import pandas as pd
from datasets import load_dataset, Dataset

# Load CSV files as Hugging Face datasets
data1 = load_dataset("csv", data_files="train_data.csv", split="train", encoding="utf-8", quotechar='"')
data2 = load_dataset("csv", data_files="train_data_tmp.csv", split="train", encoding="utf-8", quotechar='"')

# Convert to pandas DataFrames
df1 = data1.to_pandas()
df2 = data2.to_pandas()

# Optional: reset index to avoid extra columns later
df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)

# Compute rows only in df1
diff_df = df1.merge(df2, how='outer', indicator=True)
diff_df = diff_df[diff_df['_merge'] == 'left_only']
diff_df = diff_df.drop(columns=['_merge']).reset_index(drop=True)  # reset index again after filtering

# Convert back to Hugging Face dataset
diff_dataset = Dataset.from_pandas(diff_df)

# Remove any leftover index column that might have been added
if "__index_level_0__" in diff_dataset.column_names:
    diff_dataset = diff_dataset.remove_columns("__index_level_0__")

# Save to CSV
diff_dataset.to_csv("diff.csv")
```
