```python
from datasets import load_dataset

# Load CSV
dataset = load_dataset("csv", data_files="combined.csv", split="train", encoding="utf-8", quotechar='"')

def remove_brackets(batch):
    # Only process rows where Task == 'detection'
    answers = batch['Answer']
    instructions = batch['Instruction']
    tasks = batch['Task']
    new_answers = []
    new_instructions = []
    for ans, ins, task in zip (answers, instructions, tasks):
        if task == 'detection':
            ans = ans.replace('[', '').replace(']', '')
        elif task == 'replacement':
            ins = ins.replace('[', '').replace(']', '')
        new_answers.append(ans)
        new_instructions.append(ins)
    batch['Answer'] = new_answers
    batch['Instruction'] = new_instructions
    return batch
    
dataset = dataset.map(remove_brackets,batched=True)
dataset.to_csv('combined2.csv')
```
