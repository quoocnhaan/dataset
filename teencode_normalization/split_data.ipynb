```python
from datasets import load_dataset

# Load CSV as a Dataset
dataset = load_dataset("csv", data_files="train_data.csv", split="train", encoding="utf-8", quotechar='"')

def normalize(batch):
    ori_filtered = []
    nor_filtered = []
    for ori, nor in zip(batch['original'], batch['normalized']):
        if ori is None or nor is None:
            print('None')
        if len(ori.split()) == len(nor.split()):
            ori_filtered.append(ori)
            nor_filtered.append(nor)
    return {"original": ori_filtered,"normalized": nor_filtered}
balance_dataset = dataset.map(normalize, batched=True)
balance_dataset.to_csv('balance.csv')
```
